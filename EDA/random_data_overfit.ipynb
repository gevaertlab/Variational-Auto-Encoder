```python
import numpy as np
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score
import pandas as pd
```

```python
train_array = np.random.normal(size=(200, 50))
val_array = np.random.normal(size=(50, 50))
```

```python
y_train = np.random.normal(size=(200, 1))
y_val = np.random.normal(size=(50, 1))
```

```python
rfr = RandomForestRegressor()
rfr.fit(train_array, y_train)
```

```python
rfr.score(train_array, y_train)
```

```python
rfr.score(val_array, y_val)
```

```python
from scipy.stats import pearsonr
```

```python
rs, ps = [],  []
for i in range(train_array.shape[1]):
    r, p = pearsonr(train_array[:,i], y_train[:,0])
    rs.append(r)
    ps.append(p)
df = pd.DataFrame({"r":rs, "p":ps})
```
